```python
from PIL import Image
import pytesseract
import matplotlib.pyplot as plt
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Upload image
from google.colab import files
uploaded = files.upload()

# Load image
image_path = list(uploaded.keys())[0]
img = Image.open(image_path)

# REAL OCR
text = pytesseract.image_to_string(img)
print("Extracted Text:\n")
print(text)

# Sentiment Analysis
sid = SentimentIntensityAnalyzer()
scores = sid.polarity_scores(text)

neg = scores['neg'] * 100
neu = scores['neu'] * 100
pos = scores['pos'] * 100

print("\nSentiment Scores:")
print("Negative:", neg)
print("Neutral :", neu)
print("Positive:", pos)

if scores['compound'] >= 0.05:
    print("Overall Sentiment: POSITIVE")
elif scores['compound'] <= -0.05:
    print("Overall Sentiment: NEGATIVE")
else:
    print("Overall Sentiment: NEUTRAL")

# Graph
labels = ['Negative', 'Neutral', 'Positive']
values = [neg, neu, pos]

plt.bar(labels, values)
plt.title("Sentiment Analysis Result")
plt.ylabel("Percentage")
plt.show()
```

```python
import streamlit as st
from PIL import Image
import pytesseract
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
from pdf2image import convert_from_bytes

st.set_page_config(page_title="OCR & Sentiment App", layout="centered")

st.title("ðŸ“„ Text Detection & Sentiment Analysis App")

sid = SentimentIntensityAnalyzer()

uploaded_file = st.file_uploader(
    "Upload an Image / PDF / Text File",
    type=["png", "jpg", "jpeg", "pdf", "txt"]
)

extracted_text = ""

if uploaded_file:
    file_type = uploaded_file.name.split(".")[-1].lower()

    # TXT FILE
    if file_type == "txt":
        extracted_text = uploaded_file.read().decode("utf-8")

    # IMAGE FILE
    elif file_type in ["png", "jpg", "jpeg"]:
        image = Image.open(uploaded_file)
        extracted_text = pytesseract.image_to_string(image)
        st.image(image, caption="Uploaded Image", use_column_width=True)

    # PDF FILE
    elif file_type == "pdf":
        images = convert_from_bytes(uploaded_file.read())
        for img in images:
            extracted_text += pytesseract.image_to_string(img)

    st.subheader("ðŸ“Œ Extracted Text")
    st.text_area("", extracted_text, height=200)

    if st.button("Analyze Sentiment"):
        scores = sid.polarity_scores(extracted_text)

        neg = scores['neg'] * 100
        neu = scores['neu'] * 100
        pos = scores['pos'] * 100

        st.write("### Sentiment Scores")
        st.write(f"ðŸ”´ Negative: {neg:.2f}%")
        st.write(f"ðŸŸ¡ Neutral: {neu:.2f}%")
        st.write(f"ðŸŸ¢ Positive: {pos:.2f}%")

        if scores['compound'] >= 0.05:
            st.success("Overall Sentiment: POSITIVE ðŸ˜Š")
        elif scores['compound'] <= -0.05:
            st.error("Overall Sentiment: NEGATIVE ðŸ˜ž")
        else:
            st.warning("Overall Sentiment: NEUTRAL ðŸ˜")

        # Graph
        fig, ax = plt.subplots()
        ax.bar(["Negative", "Neutral", "Positive"], [neg, neu, pos])
        ax.set_ylabel("Percentage")
        ax.set_title("Sentiment Analysis Result")
        st.pyplot(fig)
```
